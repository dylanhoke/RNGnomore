```python
import random


class Character:

    def __init__(self,
                 strength,
                 intelligence,
                 dexterity,
                 endurance,
                 spirit,
                 constitution,
                 stamina,
                 health,
                 mana,
                 armor,
                 magic_resistance,
                 backstab_damage,
                 critical_damage,
                 critical_chance,
                 melee_damage,
                 ranged_damage,
                 magic_damage,
                 ambush_damage,
                 sneakattack_damage,
                 mercilessattack_damage,
                 stamina_protection,
                 parry_proficiency):
        
        self.strength = strength
        self.intelligence = intelligence
        self.dexterity = dexterity
        self.endurance = endurance
        self.spirit = spirit
        self.constitution = constitution
        self.stamina = stamina
        self.health = health
        self.mana = mana
        self.armor = armor
        self.magic_resistance = magic_resistance
        self.backstab_damage = backstab_damage
        self.critical_damage = critical_damage
        self.critical_chance = critical_chance
        self.melee_damage = melee_damage
        self.ranged_damage = ranged_damage
        self.magic_damage = magic_damage
        self.ambush_damage = ambush_damage
        self.sneakattack_damage = sneakattack_damage
        self.mercilessattack_damage = mercilessattack_damage
        self.stamina_protection = stamina_protection
        self.parry_proficiency = parry_proficiency

    def populateStats(self, strength: int = 5,
                      intelligence: int = 5,
                      dexterity: int = 5,
                      endurance: int = 5,
                      spirit: int = 5,
                      constitution: int = 5,
                      stamina: int = 100,
                      health: int = 100,
                      mana: int = 100,
                      armor: int = 50,
                      magic_resistance: int = 50,
                      backstab_damage: int = 50,
                      critical_damage: int = 50,
                      critical_chance: int = 50,
                      melee_damage: int = 25,
                      ranged_damage: int = 25,
                      magic_damage: int = 25,
                      ambush_damage: int = 0,
                      sneakattack_damage: int = 900,
                      mercilessattack_damage: int = 500,
                      stamina_protection: int = 0,
                      parry_proficiency: int = 0):

        self.strength = strength
        self.intelligence = intelligence
        self.dexterity = dexterity
        self.endurance = endurance
        self.spirit = spirit
        self.constitution = constitution
        self.stamina = stamina
        self.health = health
        self.mana = mana
        self.armor = armor
        self.magic_resistance = magic_resistance
        self.backstab_damage = backstab_damage
        self.critical_damage = critical_damage
        self.critical_chance = critical_chance
        self.melee_damage = melee_damage
        self.ranged_damage = ranged_damage
        self.magic_damage = magic_damage
        self.ambush_damage = ambush_damage
        self.sneakattack_damage = sneakattack_damage
        self.mercilessattack_damage = mercilessattack_damage
        self.stamina_protection = stamina_protection
        self.parry_proficiency = parry_proficiency

    def damage_physical(self, damage):

        damage = random.randint(0, self.health)

        if damage < self.health and (self.armor * damage) > damage:
            #survives hit
            self.health -= damage - (self.armor * damage)
        else:
            #death
            self.health = 0

    def damage_magic(self, damage):

        damage = random.randint(0, self.health)

        if damage < self.health:
            #survives hit
            self.health -= damage
        else:
            #death
            self.health = 0

```
